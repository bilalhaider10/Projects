```python
df_fake=pd.read_csv('../input/fake-and-real-news-dataset/Fake.csv')
df_true=pd.read_csv('../input/fake-and-real-news-dataset/True.csv')
```

```python
df_true['class']='true'
```

```python
frames=[df_fake,df_true]
df=pd.concat(frames)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split


train_set, test_set = train_test_split(df, test_size=.2, random_state=42)
```

```python
test_set['class'].value_counts()
```

```python
train_set['class'].value_counts()
```

```python
from fastai.text.all import *
train_set
```

```python
dls_lm = DataBlock(
    blocks=TextBlock.from_df('text', is_lm=True),
    get_x=ColReader('text'), 
    splitter=RandomSplitter(0.1) 
    # using only 10% of entire comments data for validation inorder to learn more
)
```

```python
dls_lm = dls_lm.dataloaders(train_set, bs=128,eq_len=80)
```

```python
dls_lm.show_batch(max_n=3)
```

```python
learn = language_model_learner(
    dls_lm, AWD_LSTM,drop_mult=0.3,pretrained = True,
    metrics=[accuracy, Perplexity()]).to_fp16()
```

```python
# learn.lr_find()
```

```python
learn.fit_one_cycle(1,1e-2)
```

```python
learn.save('1epoch')
```

```python
learn.unfreeze()
learn.fit_one_cycle(5)
```

```python
learn.save_encoder('finetuned')
```

```python
TEXT = "i"
N_WORDS = 10
N_SENTENCES = 2
preds = [learn.predict(TEXT, N_WORDS, temperature=0.75) 
         for _ in range(N_SENTENCES)]
```

```python
print("\n".join(preds))
```

```python
dls = TextDataLoaders.from_df(train_set, text_col='text', label_col='class',valid_pct=0.1,text_vocab=dls_lm.vocab)
```

```python
dls.show_batch(max_n=3)
```

```python
learn = text_classifier_learner(dls,AWD_LSTM,metrics=accuracy,
                                drop_mult=0.5).to_fp16()
```

```python
learn = learn.load_encoder('finetuned')
```

```python
learn.fit_one_cycle(1)
```

```python
learn.freeze_to(-2)
learn.fit_one_cycle(1)
```

```python
learn.freeze_to(-3)
learn.fit_one_cycle(1)
```

```python
learn.unfreeze()
learn.fit_one_cycle(5)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```

```python
interp.print_classification_report()
```

```python
preds,y, loss = learn.get_preds(with_loss=True)
```

```python
acc = accuracy(preds, y)
print('The accuracy is {0} %.'.format(acc))
```

```python
from sklearn.metrics import roc_curve, auc
# probs from log preds
probs = np.exp(preds[:,1])
# Compute ROC curve
fpr, tpr, thresholds = roc_curve(y, probs, pos_label=1)

# Compute ROC area
roc_auc = auc(fpr, tpr)
print('ROC area is {0}'.format(roc_auc))
```
