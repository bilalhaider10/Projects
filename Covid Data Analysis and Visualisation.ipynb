```python
#https://prsindia.org/budgets/states/jammu-and-kashmir-budget-analysis-2018-19
#https://www.statista.com/statistics/685200/india-highest-public-health-expenditure-by-state/
health=[66.68,59.03,88.16,35.29,72.3,80.65,122.25,36.38,98.14,99.76,189.67,79.21]
for i,value in enumerate(health):
    dfTotal['Health Care Spending'][i]=value
```

```python
dfTotal
```

```python
dfTotal['Health Care per capita']=None
dfTotal['population']=None
```

```python
#in millions
#https://www.indiatoday.in/education-today/gk-current-affairs/story/indian-states-with-highest-population-1358414-2018-10-08
population=[103.8,28.5,60.4,12.5,61.1,72.6,112.4,27.7,68.6,72.1,119.8,91.3]
for i,value in enumerate(population):
    dfTotal['population'][i]=value
    dfTotal['Health Care per capita'][i]=((dfTotal['Health Care Spending'][i])*1000)/(dfTotal['population'][i])
```
